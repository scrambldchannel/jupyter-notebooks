## Matplotlib

A low level library for basic, flexible plotting. It's verbose though and can require quite a lot of code to create a simple chart. 

```python
import matplotlib.pyplot as plt
```

```python
# setup some random test data

x = np.linspace(0.1, 10, 25)
y = x ** 2
```

```python
# basics

# create figure object with a size in inches
fig = plt.figure(figsize=(6,4))

# add axis to figure
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# set axes labels
ax.set_xlabel('Label for X Axis')
ax.set_ylabel('Label for Y Axis')

# plot our sample data and a label 
ax.plot(x,y, 'r-', label='X vs Y')

# add a legend (requiers a label)
ax.legend()

# Removes cell information from notebook 
plt.show()
```

```python
# using subplots

# returns a tuple of a figure and an array of axes
# note the figsize is the total size of all subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18,4))

# plot each figure
axes[0].plot(x,y, 'r')
axes[0].set_xlabel('Fig 1')
axes[1].plot(y,x, 'g')
axes[1].set_xlabel('Fig 2')
axes[2].plot(x,y * 2, 'b')
axes[2].set_xlabel('Fig 3')

# use this to attempt to fit the figures more nicely
fig.tight_layout()

plt.show()
```

```python
# multiple plots on single figure

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_xlabel('Label for X Axis')
ax.set_ylabel('Label for Y Axis')

# plot our sample data and a label 
ax.plot(x,x ** 2, 'r-', label='X²')
ax.plot(x, x ** 3, 'b-', label='X³')
# add a legend (requiers a label)
ax.legend()

plt.show()
```

```python
# Customising appearance

# multiple plots on single figure

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Customising axes appearance
ax.set_xlabel('Label for X Axis')
ax.set_ylabel('Label for Y Axis')
# set limits for axes
ax.set_xlim([0, 10])
ax.set_ylim([0, 1000])

# Customising colours and thickness
# Thick purple line
ax.plot(x, x ** 2, color='purple', lw=4, label='X²')
# Hex code color, standard thickness but transparent and with red markers
ax.plot(x, x ** 3, color='#FFCC00', marker='+', markeredgecolor='red', alpha=0.5, label='X³')

ax.legend()

plt.show()
```
